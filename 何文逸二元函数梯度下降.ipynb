```python
import numpy as np

# 定义函数 f(x, y)
def f(x, y):
    return x**3 - y**3 + 3*x**2 + 3*y**2 - 9*x

# 定义梯度函数
def gradient(x, y):
    df_dx = 3*x**2 + 6*x - 9
    df_dy = -3*y**2 + 6*y
    return df_dx, df_dy

# 梯度下降法
def gradient_descent(initial_point, learning_rate, num_iterations):
    x, y = initial_point
    for i in range(num_iterations):
        df_dx, df_dy = gradient(x, y)
        x = x - learning_rate * df_dx
        y = y - learning_rate * df_dy
        if i % 100 == 0:  # 每100次迭代打印一次结果
            print(f"Iteration {i}: x = {x}, y = {y}, f(x, y) = {f(x, y)}")
    return x, y

# 初始点、学习率和迭代次数
initial_point = (0, 0)
learning_rate = 0.01
num_iterations = 10000

# 执行梯度下降法
optimal_x, optimal_y = gradient_descent(initial_point, learning_rate, num_iterations)
print(f"Optimal point: (x, y) = ({optimal_x}, {optimal_y}), f(x, y) = {f(optimal_x, optimal_y)}")
```
